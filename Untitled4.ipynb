Datan visualisointi (T6.3 / 1p)

Visualisoi aikasarjadata haluamallasi tavalla.


Valitse tähän ainakin kaksi eri kuvaajatyyppiä

Visualisoi data kenttien 'Low price' ja 'High price' avulla


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set the figure size
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True

fig1, ax1 = plt.subplots()

df2['High price'].plot()
df2['Low price'].plot()
plt.grid(axis='both', linestyle='dashed')
plt.xlabel('date')
plt.ylabel('price')
plt.legend(["High price", "Low price"])
plt.show()
```

Visualisoi esim. käyttäen rolling-operaattorin antama tasaisesti painotettu sarja


```python

## calculating Rolling mean and storing it  
# into a new column of existing dataFram
df2 = df2['Low price'].to_frame() 
df2['M20'] = df2['Low price'].rolling(20).mean() 
#the moving average (rolling mean)  
# of 20 days 
df2
df2[['Low price', 'M20']].plot(figsize=(16, 8)) 
```

Visualisoi antaen enemmän painoa tuoreimmille havainnoille (ewm (exponentially weighted moving) -operaattori). 
Tällainen sarja adaptoituu nopeammin muutoksiin kuin rolling-operaattorin antama tasaisesti painotettu sarja.

```python
df3 = df2['Low price'].to_frame() 
# calculating exponential moving average
# using .ewm(span).mean() , with window size = 20
df3['EWMA20'] = df3['Low price'].ewm(span=20).mean()
df3

# plotting low price and exponential 
# moving averages of 20 days
# using .plot() method
df3[['Low price', 'EWMA20']].plot(figsize=(16, 8))
```

Regressiomalli (T6.5 / 1p)

Rakenna datasta Lineaarinen regressiomalli
Lineeaarinen malli riippuvuudelle saadaan tuomalla sklearn.lineal_model-kirjastosta LinearRegression-luokka
Visualisointiin voi käyttää seaborn-kirjaston regplot-metodia

Visualisointi regressiomallille

```python
from sklearn.linear_model import LinearRegression
#valitse selittävä muuttuja (x) sekä selitettävä muuttuja (y)

x=df['Opening price']
X=df['Opening price'].to_frame() #feature-matriisi
y=df['Total volume'] #target
plt.scatter(x,y)

# TODO: Regressiomalli: Lineaarinen malli valituille kentille.

malli=LinearRegression()
malli.fit(X,y)



xfit=np.linspace(2,6) 
Xfit=pd.DataFrame(xfit)
yfit=malli.predict(Xfit)
plt.scatter(x,y)
plt.plot(xfit,yfit)
plt.xlabel('Opening price')
plt.ylabel('Total volume')

# TODO: Perustele myös valinta. 
#Here we could see the predictable total volu,e depends of predictable opening price. If price grow up,total volume will decrease
```
