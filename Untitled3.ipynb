4. Autodatan esittäminen

```python
import pandas as pd
import numpy as np
import seaborn as sns
plt.rcParams['axes.facecolor'] = 'white'
# Asetetaan datalähde, josta data löytyy
url_src = "https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta5/harjoitukset/autot.csv"
df = pd.read_csv(url_src, sep='\t', decimal='.')

#count how many duplicates have each of the model
result = pd.DataFrame(df['Mh'].value_counts())
print(result)

#choose the rows where model has more than 50 diplicates
df = df[(df["Mh"] == "SKODA") | (df["Mh"] == "AUTOMOBILES CITROEN") | (df["Mh"] == "FORD WERKE GMBH") | (df["Mh"] == "NISSAN")]

# changing names of fuel type
df["Ft"] = df["Ft"].replace('DIESEL','Diesel')
df["Ft"] = df["Ft"].replace('PETROL','Petrol')
df["Ft"] = df["Ft"].replace('NG','Natural Gas')


# creating the diagram
ax = sns.countplot(y='Mh', hue='Ft', data=df)
ax.grid()
ax.set_title('Number of cars by fuel type')
ax.set_xlabel('Number of cars')
ax.set_ylabel('Manufacturer name')
ax.legend(loc='lower right', title='Fuel type')

plt.show()



```

5. Autodatan esittäminen (2)

```python
import pandas as pd
import numpy as np
import seaborn as sns
plt.rcParams['axes.facecolor'] = 'white'
# Asetetaan datalähde, josta data löytyy
url_src = "https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta5/harjoitukset/autot.csv"
df = pd.read_csv(url_src, sep='\t', decimal='.')


# change data from column first to float type and than to int
df['Ernedc (g/km)'] = pd.to_numeric(df['Ernedc (g/km)'], errors='coerce', downcast='integer')
df['Ernedc (g/km)'] = df['Ernedc (g/km)'].fillna(0).astype(int)

# choose rows where data from column more than 0 and model has more than 50 diplicates (take this data from previous exercise)
df3 = df[df['Ernedc (g/km)'] > 0]
df3 = df3[((df3["Mh"] == "SKODA") | (df3["Mh"] == "AUTOMOBILES CITROEN") | (df3["Mh"] == "FORD WERKE GMBH") | (df3["Mh"] == "NISSAN"))]

# creating diagram

ax1=sns.distplot(df3['Enedc (g/km)'].dropna(), kde=False, label=r'$Specific CO_2 Emissions $')

ax2=sns.distplot(df3['Enedc (g/km)'].dropna(), hist=False, ax=ax1, label=r'$Cum. specific CO_2 Emissions$') # cumulative histagram
ax1.set_xlabel(r'$Specific CO_2 Emissions, (g/km)$')
ax1.set_ylabel('Probabiiity density')

# Adding legend
ax1.legend(loc=0)
ax2.legend(loc=0)
ax1.legend(loc='lower right')
ax2.legend(loc='lower right')
plt.yscale('log') # skaalaus logaritmiseen asteikkoon

plt.xticks(np.arange(min(1), max(100), 1.0))

fig = plt.gcf()  #antaa nykyisen kuvion (current figure)

fig.set_size_inches(12, 3)

plt.savefig('kuvio1.png', dpi=400) # saving diagram in png format
plt.show()




 
```
